```python
# Fereydoon's Research - Applications of Deep Learning in Neuroimaging - sample script
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
anat_img = nib.load('wm20120615_111443fMRIImitation3s003a1001.nii')
anat_img_data = anat_img.get_data()
def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
show_slices([anat_img_data[35, :, :],
             anat_img_data[:, 33, :],
             anat_img_data[:, :, 38]])
print(type(anat_img_data))
plt.suptitle("Center slices for anatomical image")  
```
